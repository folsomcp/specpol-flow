# Sonifying LSD profiles

```python
import astronify as snd
import specpolFlow as pol
from astropy.table import Table
```

```python
lsd = pol.iolsd.read_lsd('SampleData/SampleLSD.s')
fig, ax = lsd.plot()
```

```python
data_table = Table({"vel":lsd.vel, "V":lsd.specV})

data_soni = snd.series.SoniSeries(data_table, time_col='vel', val_col='V')
data_soni.note_spacing = 0.02
data_soni.sonify()
data_soni.play()   
```
